```python
import requests
import json
import pandas as pd
import numpy as n
from alpha_vantage.timeseries import TimeSeries
import os
from dotenv import load_dotenv
import time
```

```python
API_KEY=None
with open(".env.txt") as f:
    for line in f:
        if line.startswith("API_KEY"):
            API_KEY=line.strip().split("=")[1]
            break
print(API_KEY)
```

```python
# ts=TimeSeries(key=API_KEY,output_format="pandas")
# data,_=ts.get_daily(symbol="MSFT",outputsize="full")
# data.shape
```

```python
all_data=[]
stocks=["ADANIENT.BSE","BAJFINANCE.BSE","APOLLOHOSP.BSE","ASIANPAINT.BSE","RELIANCE.BSE","BPCL.BSE","KOTAKBANK.BSE","TITAN.BSE"]
ts=TimeSeries(key=API_KEY,output_format="pandas")

for stock in stocks:
    data,_ = ts.get_daily(symbol=stock,outputsize="full")
    data.reset_index(inplace=True)
    data["symbol"]=stock
    all_data.append(data)
    print(f"appended stock:{stock}")
    time.sleep(15)
    
nifty_df=pd.concat(all_data,ignore_index=True)
nifty_df.to_csv("data.csv",index=False)
```

```python


ts=TimeSeries(key=API_KEY,output_format="pandas")

s=[   "BHARTIARTL.BSE","BRITANNIA.BSE", "CIPLA.BSE", "COALINDIA.BSE", "DIVISLAB.BSE", "DRREDDY.BSE",
    "HCLTECH.BSE","HEROMOTOCO.BSE","ITC.BSE","WIPRO.BSE"
]
other_data=[]
for stock in s:
    data,_=ts.get_daily(symbol=stock,outputsize='full')
    data.reset_index(inplace=True)
    data["symbol"]=stock
    other_data.append(data)
    print(f"appended stock:{stock}")
    time.sleep(15)

other_df=pd.concat(other_data,ignore_index=False)
```
